## Rosenbrock function
### $y = 100(x_{1} - x_{0}^{2})^{2} + (1 - x_{0})^{2}$

**Used when testing function optimzation because the shallow valley of this function makes hard to optimize**

```python
import numpy as np

from dezero import Variable
from dezero import Function

def rosenbrock(x0, x1):
    y = 100 * (x1 - x0**2)**2 + (1 - x0)**2
    return y
```

```python
x0 = Variable(np.array(0.0))
x1 = Variable(np.array(2.0))

y = rosenbrock(x0, x1)
y.backward()

print(x0.grad, x1.grad)
```

## Gradient descent

### Reach the optimized point `(1, 1)`

```python
def plot_rosenbrock(x0s, x1s):
    x1 = np.arange(-2, 2, 0.01)
    x2 = np.arange(-1, 3, 0.01)
    xx1, xx2 = np.meshgrid(x1, x2)

    Z = np.array([rosenbrock(x0, x1) for x0, x1 in np.vstack([xx1.ravel(), xx2.ravel()]).T]).reshape(xx1.shape)

    plt.figure(figsize=(10, 8))
    plt.contour(xx1, xx2, np.log(Z), levels=70, zorder=0)
    plt.scatter(x0s, x1s)
    plt.scatter(x0s[-1], x1s[-1], s=300, c='r', marker='^')
    plt.text(x0s[-1] + 0.1, x1s[-1], s=f'({x0s[-1]:.3f}, {x1s[-1]:.3f})')
    plt.show()
```

### iteration 1000

```python
x0 = Variable(np.array(0.0))
x1 = Variable(np.array(2.0))
lr = 0.001
iters = 1000

x0s = []
x1s = []
for i in range(iters):
    x0s.append(float(x0.data))
    x1s.append(float(x1.data))
    
    y = rosenbrock(x0, x1)
    
    x0.cleargrad()
    x1.cleargrad()
    y.backward()
    
    x0.data -= lr * x0.grad
    x1.data -= lr * x1.grad
```

```python
import matplotlib.pyplot as plt
```

```python
plot_rosenbrock(x0s, x1s)
```
